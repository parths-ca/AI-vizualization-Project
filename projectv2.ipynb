```python
def prep_time_series(data):
    try:
        data['DATE_TIME'] = pd.to_datetime(data['DATE_TIME'],format='%d-%m-%Y %H:%M')
    except:
        data['DATE_TIME'] = pd.to_datetime(data['DATE_TIME'],format='%Y-%m-%d %H:%M:%S')
    time_series_data = data.set_index('DATE_TIME')
    data_cols = ['DC_POWER','AC_POWER','DAILY_YIELD','TOTAL_YIELD']
    time_series_data_weekly = time_series_data[data_cols].resample('W').mean()
    time_series_data_daily = time_series_data[data_cols].resample('24H').mean()
    return time_series_data_weekly, time_series_data_daily
plant1_generation_weekly, plant1_generation_daily = prep_time_series(plant1_generation)
plant2_generation_weekly, plant2_generation_daily = prep_time_series(plant2_generation)
sns.set(rc={'figure.figsize':(25, 10)})
fig, ax = plt.subplots()
ax.plot(plant1_generation_weekly['AC_POWER'], marker='o',linewidth=4, linestyle='--',label='AC_POWER_WEEKLY-PLANT-1')
ax.plot(plant2_generation_weekly['AC_POWER'], marker='o',linewidth=4, linestyle='--',label='AC_POWER_WEEKLY-PLANT-2')
ax.plot(plant1_generation_daily['AC_POWER'], marker='o',linewidth=4, linestyle='--',label='AC_POWER_DAILY-PLANT-1')
ax.plot(plant2_generation_daily['AC_POWER'], marker='o',linewidth=4, linestyle='--',label='AC_POWER_DAILY-PLANT-2')
ax.set_ylabel('Solar Power Generation (kW)')
ax.legend()
```

```python
def prep_time_series_sensor(data):
    try:
        data['DATE_TIME'] = pd.to_datetime(data['DATE_TIME'],format='%d-%m-%Y %H:%M')
    except:
        data['DATE_TIME'] = pd.to_datetime(data['DATE_TIME'],format='%Y-%m-%d %H:%M:%S')
    time_series_data = data.set_index('DATE_TIME')
    data_cols = ['AMBIENT_TEMPERATURE','MODULE_TEMPERATURE','IRRADIATION']
    time_series_data_weekly = time_series_data[data_cols].resample('W').mean()
    time_series_data_daily = time_series_data[data_cols].resample('24H').mean()
    return time_series_data_weekly, time_series_data_daily
plant1_sensor_weekly, plant1_sensor_daily = prep_time_series_sensor(plant1_sensor)
plant2_sensor_weekly, plant2_sensor_daily = prep_time_series_sensor(plant2_sensor)
sns.set(rc={'figure.figsize':(25, 10)})
fig, ax = plt.subplots()
ax.plot(plant1_sensor_weekly['AMBIENT_TEMPERATURE'], marker='o',linewidth=4, linestyle='--',label='AMBIENT_TEMPERATURE_WEEKLY-PLANT-1')
ax.plot(plant2_sensor_weekly['AMBIENT_TEMPERATURE'], marker='o',linewidth=4, linestyle='--',label='AMBIENT_TEMPERATURE_WEEKLY-PLANT-2')
ax.plot(plant1_sensor_daily['AMBIENT_TEMPERATURE'], marker='o',linewidth=4, linestyle='--',label='AMBIENT_TEMPERATURE_DAILY-PLANT-1')
ax.plot(plant2_sensor_daily['AMBIENT_TEMPERATURE'], marker='o',linewidth=4, linestyle='--',label='AMBIENT_TEMPERATURE_DAILY-PLANT-2')
ax.set_ylabel('Ambient Temperature (C)')
ax.legend()
    
```

```python
sns.set(rc={'figure.figsize':(25, 10)})
fig, ax = plt.subplots()
ax.plot(plant1_sensor_weekly['AMBIENT_TEMPERATURE'] - plant1_sensor_weekly['MODULE_TEMPERATURE'], marker='o',linewidth=4, linestyle='--',label='TEMPERATURE DIFFERENCE WEEKLY PLANT-1')
ax.plot(plant2_sensor_weekly['AMBIENT_TEMPERATURE'] - plant2_sensor_weekly['MODULE_TEMPERATURE'], marker='o',linewidth=4, linestyle='--',label='TEMPERATURE DIFFERENCE WEEKLY PLANT-2')
ax.set_ylabel('Ambient Temperature (C)')
ax.legend()
    
```
